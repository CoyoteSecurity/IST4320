```python
import tkinter as tk
import sqlite3

# create database connection
conn = sqlite3.connect('userdata.db')
c = conn.cursor()

# create table if it doesn't exist
c.execute('''CREATE TABLE IF NOT EXISTS users
             (name TEXT, age INTEGER)''')
conn.commit()

# tkinter app
class UserForm(tk.Frame):
    def __init__(self, master=None):
        super().__init__(master)
        self.master = master
        self.pack()
        self.create_widgets()

    def create_widgets(self):
        # name label and entry
        self.name_label = tk.Label(self, text="Name:")
        self.name_label.pack()
        self.name_entry = tk.Entry(self)
        self.name_entry.pack()

        # age label and entry
        self.age_label = tk.Label(self, text="Age:")
        self.age_label.pack()
        self.age_entry = tk.Entry(self)
        self.age_entry.pack()

        # submit button
        self.submit_button = tk.Button(self, text="Submit", command=self.submit)
        self.submit_button.pack()

        # show data button
        self.show_button = tk.Button(self, text="Show Data", command=self.show_data)
        self.show_button.pack()

    def submit(self):
        # get user input
        name = self.name_entry.get()
        age = int(self.age_entry.get())

        # insert data into database
        c.execute("INSERT INTO users VALUES (?, ?)", (name, age))
        conn.commit()

    def show_data(self):
        # get data from database
        c.execute("SELECT * FROM users")
        rows = c.fetchall()

        # create a new window to display the data
        data_window = tk.Toplevel(self.master)
        data_window.title("User Data")
        data_frame = tk.Frame(data_window)
        data_frame.pack()

        # display data in a label
        data_label = tk.Label(data_frame, text='\n'.join([f"{row[0]}, {row[1]}" for row in rows]))
        data_label.pack()

# create tkinter app and run mainloop
root = tk.Tk()
app = UserForm(master=root)
app.mainloop()

# close database connection
conn.close()
```
